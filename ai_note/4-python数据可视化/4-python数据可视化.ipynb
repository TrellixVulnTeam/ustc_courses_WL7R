### 2.1 折线图 / '曲'线图

主要函数：

- `matplotlib.pyplot.plot`

````python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
````

````python
# 定义图
plt.figure(figsize=[12, 5])

# 定义函数
X = np.linspace(-np.pi, np.pi, 10)
Y_cos, Y_sin = np.cos(X), np.sin(X)

# 绘制函数
plt.subplot(121)
plt.plot(X, Y_cos)
plt.plot(X, Y_sin)

plt.subplot(122)
X = np.linspace(-np.pi, np.pi, 100)
Y_cos, Y_sin = np.cos(X), np.sin(X)
plt.plot(X, Y_cos)
plt.plot(X, Y_sin)

# 展示
plt.show()
````

#### 如何美化

- 颜色
- 样式
- 标题
- 坐标轴
- 脊柱
- 图例
- 注释

````python
# 定义图
plt.figure(figsize=[8, 6])

# 定义函数
X = np.linspace(-np.pi, np.pi, 50)
Y_cos, Y_sin = np.cos(X), np.sin(X)

# 绘制函数
plt.plot(X, Y_cos, color='red', marker='*', linewidth=2.5, linestyle='--', label='cosine')
plt.plot(X + 0.05, Y_sin, color='#90a4ae', marker='o', linewidth=2.5, linestyle=':', label='sin')
plt.plot(X, Y_sin, color='#00bfa5', marker='o', linewidth=2.5, linestyle=':', label='sin')

# 标题
plt.title("matplotlib demo")

# 坐标轴
plt.xlim(X.min(), X.max())
plt.ylim(-1.2, 1.2)

plt.xlabel('x label')
plt.ylabel('y label')


# 脊柱
# ax = plt.gca()
# ax.spines['right'].set_color('none')
# ax.spines['top'].set_color('none')
# ax.xaxis.set_ticks_position('bottom')
# ax.spines['bottom'].set_position(('data', 0))
# ax.yaxis.set_ticks_position('left')
# ax.spines['left'].set_position(('data', 0))

# 图例
plt.legend(loc='upper left')

# 注释
x = 2 * np.pi / 3
plt.scatter([x, ], [np.cos(x), ], 100, color='red')
plt.annotate(r'$ ( \frac{2\pi}{3}, \frac{\sqrt{3}}{2} )$',
            xy=(x, np.cos(x)),
            xycoords='data',
            xytext=(0.8, -0.8),
            fontsize=16,
            arrowprops=dict(arrowstyle='->', connectionstyle="arc3, rad=.2"))

# 背景颜色
ax = plt.gca()
ax.set_facecolor('#eceff1')

# 展示
plt.show()
````

### 2.2 子图

主要函数：

- `matplotlib.pyplot.subplot`

````python
plt.figure(figsize=[10, 8])

plt.subplot(221)  # 2行2列的第一个子图
plt.plot(X, Y_sin, color='#ec407a')
plt.title('1')
plt.fill_between(X, 0, Y_sin, color='#ec407a', alpha=0.6)

plt.subplot(222)
plt.plot(X, Y_sin)
plt.fill_between(X, 0, Y_sin, alpha=0.6)
plt.title('2')

plt.subplot(223)
plt.fill_between(X, Y_cos - 1, Y_cos + 1, color='#ec407a', alpha=0.6)
plt.title('3')

plt.subplot(224)
plt.title('4')
plt.fill_between(X, 0, Y_cos + 1, alpha=0.6)
plt.fill_between(X + 1.5, 0, Y_cos + 1, alpha=0.6)
````

````python
import matplotlib.gridspec as gridspec

plt.figure(figsize=[8, 8])
grid = gridspec.GridSpec(3, 3)

plt.subplot(grid[0, :])
plt.xticks([])
plt.yticks([])
plt.plot(X, Y_sin, color='w')
plt.fill_between(X[50:], 0, Y_sin[50:], color='#ec407a', alpha=0.6)
plt.fill_between(X[:50], 0, Y_sin[:50], alpha=0.6)
X += 1.5
plt.plot(X, Y_sin, color='w')
plt.fill_between(X[50:], 0, Y_sin[50:], color='#ec407a', alpha=0.6)
plt.fill_between(X[:50], 0, Y_sin[:50], alpha=0.6)

plt.subplot(grid[1, 0])
plt.xticks([])
plt.yticks([])
plt.fill_between(X, 0, Y_sin, color='#ec407a', alpha=0.6)

plt.subplot(grid[1, 1:])
plt.xticks([])
plt.yticks([])
plt.fill_between(X, 0, Y_cos + 1, alpha=0.6)
plt.fill_between(X + 1.5, 0, Y_cos + 1, alpha=0.6)

plt.subplot(grid[2, :2])
plt.xticks([])
plt.yticks([])
plt.fill_between(X, Y_cos, Y_cos + 1, color='#ec407a', alpha=0.6)

plt.subplot(grid[2, 2])
plt.fill_between(X, Y_cos, Y_cos + 1, alpha=0.6)
plt.xticks([])
plt.yticks([])
````

## 2.4 条形图

主要函数：

- `matplotlib.pyplot.bar`


````python
# 生成数据
N = 10
X = np.arange(N)
Y = (1. - X / float(N)) * np.random.uniform(0.5, 1.0, N)

# 条形图
plt.figure(figsize=[10, 6])
plt.bar(X, Y, facecolor='#ff9800')
plt.plot(X, Y, marker='^', markersize=15, linewidth=3)
plt.title('bar')

# 显示数据
for x, y in zip(X, Y):
    plt.text(x, y + 0.03, 
             '%.2f' % y, ha='center', va='bottom', 
             fontdict={'color': '#0091ea',
                      'size': 16})

plt.ylim(0., Y.max() + 0.1)
plt.xticks([])
plt.yticks([])
````

## 3. 图像可视化

RGB图像存储格式一般为多维数组：`H x W x 3`


### 3.1 读取图片（RGB图片为例）、显示图片
| 函数 | 备注 |
|:-|:-|
|`PIL.Image.open` | HxWx3 (RGB) (二进制编码方式读取，需要`np.array()`转化）|
|`cv2.imread`     |  HxWx3 __(BGR)__, 速度最快 |
|`matplotlib.image.imread` | HxWx3 (RGB)|
|`scipy.misc.imread (imageio.imread)` | HxWx3 (RGB)|
|`skimage.io.imread` | HxWx3 (RGB)|

````python
from PIL import Image
import cv2
import matplotlib.image as pltimg
import scipy.misc as scimisc
import imageio
import skimage.io as skiio


# PIL
img = Image.open('data/1.jpg')
img = np.array(img)
print(img.shape)

# opencv
img = cv2.imread('data/1.jpg')[:, :, ::-1]  # 需要将BGR转化为RGB
print(img.shape)

# matplotlib
img = pltimg.imread('data/1.jpg')
print(img.shape)

# scipy
img = scimisc.imread('data/1.jpg')
print(img.shape)

# imageio
img = imageio.imread('data/1.jpg')
print(img.shape)

# skimage
img = skiio.imread('data/1.jpg')
print(img.shape)

# 显示图片
plt.imshow(img)
````

### 3.2 检测框

`boundingbox`框一般出现在object detection等任务中。

````python
box = np.load('data/box.npy')
print("box:", box)
````

`box` 为框的坐标：`[[x1, y1], [x2, y2]]`

````python
#　＝＝＝＝添加box框＝＝＝＝
import matplotlib.patches as pth

# 显示原图
plt.figure(figsize=[6, 6])
plt.imshow(img, alpha=0.9)
plt.axis('off')

# 显示box框
ax = plt.gca()
rec = pth.Rectangle(box[0], 
                    box[1][0] - box[0][0], 
                    box[1][1] - box[0][1], 
                    facecolor='none', edgecolor='c', linewidth=4)
ax.add_patch(rec)

# 添加文字框
plt.text(box[0, 0], 
         box[0, 1], 
         "person", 
         size=20,
         fontdict=dict(color='w'),
         bbox=dict(edgecolor='c',
                   facecolor='c')
        )

print("image shape:", img.shape)
````

### 3.3 Mask蒙版

mask一般出现在semantic segmentation、instance segmentation等任务中，mask上不同值代表原图中不同的物体。

````python
mask = np.load('data/mask.npy')
print('shape:', mask.shape)
print('max:', mask.max())
print('min:', mask.min())
````

`mask`是二值化蒙版图片，值为`1`的区域为人所在区域，`0`的区域为背景，大小`224x200`与原图一样

````python
# 显示mask
plt.figure(figsize=[12, 6])

# default colormap
plt.subplot(131)
plt.imshow(mask)  # 显示为伪彩色
plt.axis('off')

# `gray` colormap
plt.subplot(132)
plt.imshow(mask, cmap='gray')
plt.axis('off')

plt.subplot(133)
plt.axis('off')
plt.imshow(mask, cmap='winter')
````

利用`alpha`参数,将mask蒙版叠加在原图上

````python
plt.figure(figsize=[10, 8])

plt.subplot(121)
plt.imshow(img)
plt.axis('off')

plt.subplot(122)
# 原图
plt.imshow(img, alpha=0.5)
plt.axis('off')
# mask
plt.imshow(mask, cmap='winter', alpha=0.5)
````

将原图和mask图片融合为一张图片再显示

- mask为一通道图片，首先用`cv2.applyColorMap`转化为三通道RGB图片
- 通过手动设定`alpha`参数，将两图片融合

````python
color_mask = cv2.applyColorMap((mask * 255.).astype(np.uint8), cv2.COLORMAP_SUMMER)
plt.axis('off')
plt.imshow(color_mask)
````

````python
# 融合图片
alpha = 0.65
mask_idx = (mask == 1)
img_mask = img.copy()
img_mask[mask_idx] = img_mask[mask_idx] * (1. - alpha) + color_mask[mask_idx] * alpha
````

````python
plt.figure(figsize=[10, 8])

plt.subplot(121)
plt.axis('off')
plt.imshow(img)

plt.subplot(122)
plt.axis('off')
plt.imshow(img_mask)
````

```

7 - Right elbow

10 - Left elbow
```
